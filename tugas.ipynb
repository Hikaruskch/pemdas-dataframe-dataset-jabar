```python
import pandas as pd
```

```python
df_data_sampah = pd.read_csv ('datasampah.csv')[['nama_kabupaten_kota','jumlah_produksi_sampah', 'tahun']]
df_data_sampah
```

```python
df_data_sampah = pd.read_excel ('datasampah.xlsx')
df_data_sampah
```

```python
# Dari DataFrame yang telah dibuat, hitunglah total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat 
# untuk tahun tertentu
total_produksi_sampah_pertahun = int (input('Masukkan tahun yang ingin dihitung: '))
total_produksi_sampah = 0
tahun_ditemukan = False
for i, row in df_data_sampah.iterrows():
    if row ['tahun'] == total_produksi_sampah_pertahun :
        total_produksi_sampah += row ['jumlah_produksi_sampah']
        tahun_ditemukan = True
        
if tahun_ditemukan:
    print(f'Total produksi {total_produksi_sampah_pertahun}: {total_produksi_sampah} ton perhari')
else:
    print(f'tidak ada data sampah pada tahun {total_produksi_sampah_pertahun}')
```

```python
# Jumlahkan data pertahun
total_pertahun = {}

for i, row in df_data_sampah.iterrows():
    tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    
    if pd.isna(jumlah):
        jumlah = 0
        
    total_pertahun[tahun] = total_pertahun.get(tahun, 0) + jumlah
    
print('\nTotal Produksi sampah pertahun')
for tahun, total in total_pertahun.items():
    print(f'Pada tahun ke-{tahun}: {total} Ton Perhari')
```

```python
# Jumlahkan data per Kota/Kabupaten per tahun
total_per_daerah = {}
for i, row in df_data_sampah.iterrows():
    wilayah = row['nama_kabupaten_kota']
    jumlah = row['jumlah_produksi_sampah']
    
    if pd.isna(jumlah):
        jumlah = 0
        
    total_per_daerah[wilayah] = total_per_daerah.get(wilayah,0) + jumlah
    
print('\nTotal Produksi sampah per wilayah')
for wilayah, total in total_per_daerah.items():
    print(f'Pada wilayah {wilayah}: {total:.2f} Ton Perhari')
```
